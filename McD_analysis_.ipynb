```python
# creating a list with our ranked list by Calories
lists=["Chicken & Fish","Breakfast","Smoothies & Shakes","Beef & Pork","Snacks & Sides","Desserts","Beverages"]
```

### *Using some libraries, the following lines of code displays an analysis of the most ranked items : -*

```python
for list in lists:
    #items ascending=False
    items=data.query('Category == list').sort_values(by="Calories", ascending=False)
    items_custom=items.iloc[:,[1,3,5]]
    
    #items ascending=True
    items_t=data.query('Category == list').sort_values(by="Calories", ascending=True)
    items_custom_t=items_t.iloc[:,[1,3,5]]
    
    #Print Categories
    print(lists.index(list)+1,".",list)
    p_express = px.scatter(items_custom, x="Total Fat",y='Calories', color="Item",size='Calories',template="plotly_dark",hover_name="Item")
    p_express.show()
    
    #Display Tables ranked by calories
    table=items_custom.iloc[:,[0,1]].sort_values(by="Calories", ascending=False).style.hide_index()
    #table=table.style.hide_index()
    display(table)
    
    #Matplotlib pyplot
    fig,ax=plt.subplots()
    x=items_custom_t.iloc[:,0]
    y=items_custom_t.iloc[:,1]
    ax.barh(x,y, tick_label=x)
    plt.show()
```

### *The following items are evaluated on the basis on Sugars, Vitamins (A&C), Calcium, Iron.*


```python
breakfast=data[data['Category']=='Breakfast']
cols=['Sugars','Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)','Iron (% Daily Value)']
for col in cols:
    print(col)
    a=breakfast.groupby('Item')[col].max().sort_values(ascending=False).head(1)
    print(a)
    print('-'*40)
    print('\n')
```
